```python
import pandas as pd
import numpy as np
```

## 参数设置
结果显示signalN = 2，BeginYear = 2006时，动量策略超额收益最大

```python
##%% 重要参数设置
##%% 设置用多少个月的收益率作动量信号
signalN = 2
##%% 设置从哪一年开始(最小2003),测试结果显示当Signal=2时，2003-2005年超额为负，2006以后为正
BeginYear = 2006
```

```python
bond1 = pd.read_excel('中债-国开行债券总财富(1-3年)指数.xlsx',usecols=[2,6],index_col=0)['2003':]
bond5 = pd.read_excel('中债-国开行债券总财富(3-5年)指数.xlsx',usecols=[2,6],index_col=0)['2003':]
bond10 = pd.read_excel('中债-国开行债券总财富(7-10年)指数.xlsx',usecols=[2,6],index_col=0)['2003':]
```

```python
bond1.head()
```

```python
bonds = pd.concat([bond1, bond5, bond10],axis = 1)
bonds.columns = ['1year','5year','10year']
```

```python
bonds.plot()
```

```python
monthR = bonds/bonds.shift(1)-1
monthR = monthR[str(BeginYear):].fillna(0)
signal = bonds/bonds.shift(signalN)-1
signal = signal.shift(1)[str(BeginYear):].fillna(0)
```

```python
monthR.plot()
```

```python
shortSignal = (signal['10year']<signal['1year'])
longSignal = (signal['10year']>signal['1year'])
meanSignal = (signal['10year']==signal['1year'])
momentumR = longSignal*monthR['10year']+shortSignal*monthR['1year']+meanSignal*(monthR['10year']+monthR['1year'])*0.5
momentumR = pd.DataFrame(momentumR.fillna(0),columns=['momentum'])
meanWeightR = pd.DataFrame(monthR[['1year','10year']].mean(axis=1),columns = ['1&10meanWeight'])
momentumAllR = pd.concat([monthR[['1year','10year']],meanWeightR,momentumR],axis=1)
momentumAll = ((momentumAllR+1).cumprod())*100
```

```python
momentumAll.plot()
```

```python
momentumAll[['momentum','1&10meanWeight']].plot()
```

```python
TrueORnot = (momentumAllR['momentum']-momentumAllR['1&10meanWeight'])>=0
```

```python
pd.concat([momentumAllR[['1year','10year','1&10meanWeight','momentum']],TrueORnot],axis=1)
```

```python
## 策略有效的频率（月超额收益>=0）
TrueORnot.sum()/len(TrueORnot)
```

```python
## 策略完全有效的频率（月超额收益>0）
BiggerORnot = (momentumAllR['momentum']-momentumAllR['1&10meanWeight'])>0
BiggerORnot.sum()/len(BiggerORnot)
```

```python
yearExcess = pd.DataFrame(index = np.arange(BeginYear,2020),columns = ['ExcessReturn'])
for year in range(BeginYear,2020):
    MonthReturn = momentumAllR[str(year)][['1&10meanWeight','momentum']]
    YearReturn = (MonthReturn+1).prod()-1
    yearExcess['ExcessReturn'][year]=YearReturn['momentum']-YearReturn['1&10meanWeight']
```
