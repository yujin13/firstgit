```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
##%% 重要参数设置
##%% 设置用多少个月的收益率作凸度信号
signalN = 1
##%% 设置从哪一年开始(最小2003)
BeginYear = 2003
## 设定多空临界值
door = 0.0
```

```python
bond1 = pd.read_excel('中债-国开行债券总财富(1-3年)指数.xlsx',usecols=[2,6],index_col=0)['2003':]
bond5 = pd.read_excel('中债-国开行债券总财富(3-5年)指数.xlsx',usecols=[2,6],index_col=0)['2003':]
bond10 = pd.read_excel('中债-国开行债券总财富(7-10年)指数.xlsx',usecols=[2,6],index_col=0)['2003':]
```

```python
bonds = pd.concat([bond1, bond5, bond10],axis = 1)
bonds.columns = ['1year','5year','10year']
```

```python
monthR = bonds/bonds.shift(1)-1
monthR = monthR[str(BeginYear):].fillna(0)
signal = bonds/bonds.shift(signalN)-1
signal = signal.shift(1)[str(BeginYear):].fillna(0)
```

```python
CurveSignal = signal['1year']+signal['10year']-2*signal['5year']
```

```python
CurveSignal.head()
```

```python
LongDivShort = monthR['10year']-monthR['1year']
```

```python
plt.plot(list(CurveSignal),list(LongDivShort),'*')
```

###  策略描述：
curve>=阈值时，做多短债
curve<阈值时，做多长债

```python
shortSignal = (CurveSignal>=door)
longSignal = (CurveSignal<door)
curveR = longSignal*monthR['10year']+shortSignal*monthR['1year']
curveR = pd.DataFrame(curveR.fillna(0),columns=['curve'])
meanWeightR = pd.DataFrame(monthR[['1year','10year']].mean(axis=1),columns = ['1&10meanWeight'])
curveAllR = pd.concat([monthR[['1year','10year']],meanWeightR,curveR],axis=1)
curveAll = ((curveAllR+1).cumprod())*100
```

```python
curveAll.plot()
```

```python
curveAll[['1&10meanWeight','curve']].plot()
```

```python
yearExcess = pd.DataFrame(index = np.arange(BeginYear,2020),columns = ['ExcessReturn'])
for year in range(BeginYear,2020):
    MonthReturn = curveAllR[str(year)][['1&10meanWeight','curve']]
    YearReturn = (MonthReturn+1).prod()-1
    yearExcess['ExcessReturn'][year]=YearReturn['curve']-YearReturn['1&10meanWeight']
```

```python
yearExcess.plot(kind = 'bar')
```
