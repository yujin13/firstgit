```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
##%% 重要参数设置
##%% 设置用多少个月的收益率作信号
signalN1 = 2#动量
signalN2 = 3#股市
signalN3 = 1#凸度
##%% 设置从哪一年开始(最小2003),测试结果显示当Signal=2时，2003-2005年超额为负，2006以后为正
BeginYear = 2006
## 设定多空临界值
door = 0.0
```

```python
bond1 = pd.read_excel('中债-国开行债券总财富(1-3年)指数.xlsx',usecols=[2,6],index_col=0)['2003':]
bond5 = pd.read_excel('中债-国开行债券总财富(3-5年)指数.xlsx',usecols=[2,6],index_col=0)['2003':]
bond10 = pd.read_excel('中债-国开行债券总财富(7-10年)指数.xlsx',usecols=[2,6],index_col=0)['2003':]
hs300 = pd.read_excel('沪深300月度2002起.xlsx',usecols=[2,6],index_col=0)['2003':][:-1]
```

```python
hs300.index = bond1.index
bonds = pd.concat([bond1, bond5, bond10, hs300],axis = 1)
bonds.columns = ['1year','5year','10year','hs300']
bonds = bonds[:-1]
```

```python
monthR = bonds/bonds.shift(1)-1
monthR = monthR[str(BeginYear):].fillna(0)
#动量
signal1 = bonds/bonds.shift(signalN1)-1
signal1 = signal1.shift(1)[str(BeginYear):].fillna(0)
#股市
signal2 = bonds['hs300']/bonds['hs300'].shift(signalN2)-1
signal2 = signal2.shift(1)[str(BeginYear):].fillna(0)
#凸度
signal3 = bonds/bonds.shift(signalN3)-1
signal3 = signal3.shift(1)[str(BeginYear):].fillna(0)
CurveSignal = signal3['1year']+signal3['10year']-2*signal3['5year']
LongDivShort = monthR['10year']-monthR['1year']
```

```python
#动量
shortSignal1 = (signal1['10year']<=signal1['1year'])
longSignal1 = (signal1['10year']>signal1['1year'])
#股市
shortSignal2 = (signal2>=door)
longSignal2 = (signal2<door)
#凸度
shortSignal3 = (CurveSignal>=door)
longSignal3 = (CurveSignal<door)
#合成
shortSignal = shortSignal1+shortSignal2+shortSignal3
longSignal = longSignal1+longSignal2+longSignal3

SignalsAddR = longSignal*monthR['10year']+shortSignal*monthR['1year']
SignalsAddR = pd.DataFrame(SignalsAddR.fillna(0),columns=['SignalsAdd'])
meanWeightR = pd.DataFrame(monthR[['1year','10year']].mean(axis=1),columns = ['1&10meanWeight'])
SignalsAddAllR = pd.concat([monthR[['1year','10year']],meanWeightR,SignalsAddR],axis=1)
SignalsAddAll = ((SignalsAddAllR+1).cumprod())*100
```

```python
SignalsAddAll.plot()
```

```python
SignalsAddAll[['1&10meanWeight','SignalsAdd']].plot()
```

```python
yearExcess = pd.DataFrame(index = np.arange(BeginYear,2020),columns = ['ExcessReturn'])
for year in range(BeginYear,2020):
    MonthReturn = SignalsAddAllR[str(year)][['1&10meanWeight','SignalsAdd']]
    YearReturn = (MonthReturn+1).prod()-1
    yearExcess['ExcessReturn'][year]=YearReturn['SignalsAdd']-YearReturn['1&10meanWeight']
```
